This Jupyter Notebook encompasses all the preprocessing of the images prior to using different methods of style transfer.

```python
import numpy as np
import matplotlib.pyplot as plt

import cv2
import os
```

The images are located in the directory `./icons50/icons-50/Icons-50/`

```python
def load_process_image(img, shape=(120, 120)):
    image = cv2.imread(img)
    image = cv2.resize(image, (120, 120))
    edges = cv2.Canny(image, 100, 200)
    edges = 255 - edges
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB), edges
```

```python
IMG_LOCATION = './icons50/icons-50/Icons-50/'
OUT_DIR = './data/'
i = 1
for folder in os.listdir(IMG_LOCATION):
    if not os.path.isfile(IMG_LOCATION + folder):
        for image in os.listdir(IMG_LOCATION + folder):
            img, edges = load_process_image(f'{IMG_LOCATION}{folder}/{image}')
            plt.imsave(f"{OUT_DIR}icons/{i}.png", img)
            plt.imsave(f"{OUT_DIR}edges/{i}.png", edges, cmap='gray')
            i += 1
```
